```python
import pandas as pd
import seaborn as sns
```

```python
vendas = pd.read_excel('varejo.xlsx')
vendas.head()
```

```python
vendas.info()
```

```python
vendas.groupby("idcanalvenda").idcompra.nunique()
```

```python
vendas["idcanalvenda"] = vendas['idcanalvenda'].str.replace("APP","Aplicativo")
vendas.groupby('idcanalvenda').idcompra.nunique()
```

```python
vendas.groupby("bandeira").idcompra.nunique()
```

```python
vendas.groupby("Nome_Departamento").idcompra.nunique()
```

```python
vendas.groupby("estado").idcompra.nunique()
```

```python
vendas["Nome_Departamento"] = vendas["Nome_Departamento"].str.replace(" ","_")
vendas.groupby('Nome_Departamento').idcompra.nunique()
```

```python
vendas[vendas['estado'].isnull()]
```

```python
vendas['estado'].fillna("MS", inplace=True)
vendas.groupby('estado').idcompra.nunique()
```

```python
vendas.info()
```

```python
media_preco = vendas["Preço"].mean()
media_preco
```

```python
vendas['Preço'].fillna(media_preco,inplace=True)
vendas.info()
```

```python
vendas_correta = vendas.query("Preço<Preço_com_frete")
vendas_correta.head()
```

```python
vendas_correta.describe()
```

```python
vendas_correta.groupby("Nome_Departamento").idcompra.nunique().sort_values(ascending=False).reset_index()
```

```python
round(vendas_correta.groupby("Nome_Departamento")['Preço_com_frete'].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
vendas_correta.groupby("Data").idcompra.nunique().sort_values(ascending=False)
```

```python
vendas_correta["Mes"]=vendas_correta["Data"].dt.month
vendas_correta.head()
```

```python
vendas_correta.groupby("Mes").idcompra.nunique().sort_values(ascending=False)
```

```python
cliente = pd.read_excel('cliente_varejo.xlsx')
cliente.head()
```

```python
cliente.info()
```

```python
cliente[['idade','renda']].describe()
```

```python
cliente = cliente.astype({'renda':'float'})
cliente.info()
```

```python
vendas_cliente = vendas_correta.merge(cliente,how='left',on='cliente_Log')
vendas_cliente.head()
```

```python
agg_canal_venda = round(vendas_cliente.groupby('idcanalvenda')['renda'].agg('mean').sort_values(ascending=False).reset_index())
agg_canal_venda
```

```python
agg_bandeira = vendas_cliente.groupby('bandeira')['idade'].agg('mean').sort_values(ascending=False).reset_index()
```

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(5,5))
plt.bar(agg_bandeira['bandeira'],agg_bandeira['idade'],color='green')
plt.ylabel("Média de Idade",fontweight='bold')
plt.xlabel('Bandeira',fontweight='bold')
plt.title('Média de idade por bandeira', loc='center',fontsize=18,fontweight='bold',color='black')
```

```python
fig = plt.figure(figsize=(5,5))
graf2=plt.bar(agg_canal_venda['idcanalvenda'],agg_canal_venda['renda'],color='green')
plt.ylabel('Renda')
plt.xlabel('Canal de Venda')
plt.title('Média de Renda por Canal de Venda',fontweight='bold')
plt.bar_label(graf2)
```

```python
venda_por_data = vendas_correta.groupby("Data").idcompra.nunique().reset_index()
venda_por_data
```

```python
fig = plt.figure(figsize=(10,7))
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.plot(venda_por_data['Data'],venda_por_data['idcompra'])
```

```python
vendas_departamento = round(vendas_correta.groupby("Nome_Departamento")['Preço_com_frete'].agg("mean").sort_values(ascending=False).reset_index(),2)
```

```python
fig=plt.figure(figsize=(30,10))
plt.bar(vendas_departamento['Nome_Departamento'],vendas_departamento['Preço_com_frete'],color='green')
```
